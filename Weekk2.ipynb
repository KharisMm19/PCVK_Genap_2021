```
#1. Tampilkan image dalam channel Red-Green dan Green-Blue saja!
img_rg = img.copy()
img_rg[:,:,2] = 0
plt.imshow(img_rg)
```

```
img_gb = img
img_gb[:,:,0] = 0
plt.imshow(img_gb)
```

```
#2. Tampilkan image baris ke 10-100, kolom 10-100!
plt.imshow(img[10:100,10:100])
```

```
#3. Tampilkan image baris ke 5-30, semua kolom, channel Green saja!
plt.imshow(img[5:30])
```

```
#4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!
img_rectangel = cv.cvtColor(img,cv.COLOR_BGR2RGB)
cv.rectangle(img_rectangel, pt1=(1300,500), pt2=(1500,1100), color=(0,50,255), thickness=20)
cv.rectangle(img_rectangel, pt1=(700,1200), pt2=(1500,1500), color=(0,255,100),thickness=25)
cv.rectangle(img_rectangel, pt1=(300,1000), pt2=(1100,400), color=(100,150,255),thickness=30)
cv.rectangle(img_rectangel, pt1=(600,1200), pt2=(1200,500), color=(255,0,222),thickness=35)
cv.rectangle(img_rectangel, pt1=(2400,250), pt2=(2000,1300), color=(155,0,50),thickness=40)
plt.imshow(img_rectangel)
```

```
#5. Tampilkan image dengan posisi terbalik!

img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kitten1.jpg')
img = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB),0)
plt.imshow(img)
```

```
#6. Buat rectangle dan circle pada bagian wajah dari image kitten!

cv.rectangle(img, pt1=(1200,1300), pt2=(1700,1800), color=(255,0,255), thickness=20)
cv.circle(img, center=(1450, 1550), radius=250, color=(255,0,0), thickness=30)
plt.imshow(img)
```
